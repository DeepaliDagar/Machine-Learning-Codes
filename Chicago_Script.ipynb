```python
# Link to Google Collab: https://colab.research.google.com/drive/1m0FsXdFjAYWUmHMBIoiJps8o7NDKG4pt?usp=sharing
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ingest data from a CSV file
df = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/CarPrice.csv', sep=',')
```

```python
#looking at the data
df.head(5)
```

```python
# Use a function for data analysis
def analyze_data(df):
    # Perform data analysis here

    for c in df.columns:
      print(c,':dtype:',df[c].dtype,',null count=',df[c].isna().sum())

    return
```

```python
analyze_data(df)
```

```python
# Visualize data
plt.figure(figsize=(8, 5))
plt.scatter(df['enginesize'], df['horsepower'])
plt.xlabel('Engine Size')
plt.ylabel('Horse Power')
plt.title('Engine size and correspinding Horse Power')
plt.savefig('enginesize_power.png')
```

```python
car_count = df['fueltype'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 4))
car_count.plot(kind='bar', color='blue')

# Adding labels and a title
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.title('Count of cars w.r.t fuel type')
plt.savefig('fuel_type.png')
```

```python
# Wrangle data
df.drop(columns=['symboling','aspiration','drivewheel'],inplace=True)
df.rename(columns={'highwaympg':'highway_mpg','citympg':'city_mpg'},inplace=True)
```

```python
#Defining a function
def describe_data(data):

    data = data.select_dtypes(include=['float', 'int'])
    descriptions = []

    for column in data.columns:
        description = {
            "Column Name": column,
            "Mean": data[column].mean(),
            "Median": data[column].median(),
            "Minimum": data[column].min(),
            "Maximum": data[column].max(),
            "Standard Deviation": data[column].std(),

        }
        descriptions.append(description)

    return pd.DataFrame(descriptions)
```

```python
describe_data(df)
```
